Fresnel's Equation show that the amplitude coefficiens depend on the value of the angle of incidence $\theta$.

$ Fresnel's $ $ Equation: $

For $s -$ $polarized$ $wave:$

(Elecric Vector is perpendicular to the plane of incidence)
$$ r_\perp = \frac{E_{10}}{E_0} = \frac{n_1 cos\theta - n_2 cos\theta_2}{n_1 cos\theta + n_2 cos\theta_2} $$
$$ t_\perp = \frac{E_{20}}{E_0} = \frac{2n_1 cos\theta}{n_1 cos\theta + n_2 cos\theta_2} $$

For $p -$ $polarized$ $wave:$

(Elecric Vector is parallel to the plane of incidence)
$$ r_\parallel = \frac{E_{10}}{E_0} = \frac{n_2 cos\theta - n_1 cos\theta_2}{n_2 cos\theta + n_1 cos\theta_2} $$
$$ t_\parallel = \frac{E_{20}}{E_0} = \frac{2n_1 cos\theta}{n_2 cos\theta + n_1 cos\theta_2} $$

```python
import numpy as np
import matplotlib.pyplot as plt
n1, n2 = 1, 1.5
theta = np.arange(0, np.pi/2, 0.0001)

def r_per(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return ( n1*np.cos(x1) - n2*np.cos(x2) )/( n1*np.cos(x1) + n2*np.cos(x2) )
def r_para(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(theta)*n1/n2)
  return ( n2*np.cos(x1) - n1*np.cos(x2) )/( n2*np.cos(x1) + n1*np.cos(x2) )
def t_per(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return ( 2*n1*np.cos(x1) )/(n1*np.cos(x1) + n2*np.cos(x2))
def t_para(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return ( 2*n1*np.cos(x1) )/(n2*np.cos(x1) + n1*np.cos(x2))

plt.plot(theta, r_per(theta), label=r'$r_\perp$')
plt.plot(theta, r_para(theta), label=r'$r_\parallel$')
plt.plot(theta, t_per(theta), label=r'$t_\perp$')
plt.plot(theta, t_para(theta), label=r'$t_\parallel$')
#Brewster angle
for i in theta:
  if 0.0-0.00001 < r_para(i) < 0.00001:
    plt.plot(i, r_para(i), 'ro')
    plt.text(i, 0.1, r'$\theta_p$')
plt.axhline(color='k',ls='--')
plt.ylim(-1.0,1.0)
plt.legend()
plt.xlabel(r'$\theta$ in radians')
plt.ylabel('Amplitude coefficients')
plt.show()
```

## Total Internal Refelction

$n_1 > n_2$

```python
n1, n2 = 1.5, 1
def r_per(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return ( n1*np.cos(x1) - n2*np.cos(x2) )/( n1*np.cos(x1) + n2*np.cos(x2) )
def r_para(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(theta)*n1/n2)
  return ( n2*np.cos(x1) - n1*np.cos(x2) )/( n2*np.cos(x1) + n1*np.cos(x2) )
plt.plot(theta, r_per(theta), label=r'$r_\perp$')
plt.plot(theta, r_para(theta), label=r'$r_\parallel$')
#Brewster angle
for i in theta:
  if 0.0-0.00001 < r_para(i) < 0.00001:
    plt.plot(i, r_para(i), 'ro')
    plt.text(i, 0.1, r'$\theta_p$')
#Critical angle
for i in theta:
  if 0.95 < r_para(i) and 0.95 < r_per(i):
    plt.plot(i, 0, 'ro')
    plt.text(i, 0.1, r'$\theta_c$')
plt.axhline(color='k',ls='--')
plt.ylim(-1.0,1.0)
plt.xlim(0,np.pi/2)
plt.legend()
plt.xlabel(r'$\theta$ in radians')
plt.ylabel('Amplitude coefficients')
plt.show()
```

### Reflectance and Transmittance

$$ R_\parallel = \left( \frac{n_2 cos\theta - n_1cos\theta_2}{n_2 cos\theta + n_1cos\theta_2} \right)^2 $$

$$ T_\parallel = \left( \frac{2n_1 cos\theta}{n_2 cos\theta + n_1cos\theta_2} \right)^2 . \frac{n_2cos\theta_2}{n_1cos\theta} $$

and

$$ R_\perp = \left( \frac{n_1 cos\theta - n_2cos\theta_2}{n_1 cos\theta + n_2cos\theta_2} \right)^2 $$

$$ T_\perp = \left( \frac{2n_1 cos\theta}{n_1 cos\theta + n_2cos\theta_2} \right)^2 . \frac{n_2cos\theta_2}{n_1cos\theta} $$

```python
theta = np.arange(0, np.pi/2, 0.0001)

def R_per(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return (( n1*np.cos(x1) - n2*np.cos(x2) )/( n1*np.cos(x1) + n2*np.cos(x2) ))**2
def R_para(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return (( n2*np.cos(x1) - n1*np.cos(x2) )/( n2*np.cos(x1) + n1*np.cos(x2) ))**2
def T_per(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return (( 2*n1*np.cos(x1) )/(n1*np.cos(x1) + n2*np.cos(x2)))**2 * ( (n2*np.cos(x2))/(n1*np.cos(x1)) )
def T_para(theta):
  x1 = theta
  x2 = np.arcsin(np.sin(x1)*n1/n2)
  return (( 2*n1*np.cos(x1) )/(n2*np.cos(x1) + n1*np.cos(x2)))**2 * ( (n2*np.cos(x2))/(n1*np.cos(x1)) )

plt.plot(theta, R_per(theta), label=r'$R_\perp$')
plt.plot(theta, T_per(theta), label=r'$T_\perp$')
#plt.axhline(1.0,color='k',ls='--')
#plt.axvline(np.pi/2,color='k',ls='--')
#plt.ylim(0.0,1.0)
plt.legend()
plt.xlabel(r'$\theta$ in radians')
plt.ylabel(r'$R_\perp and T_\perp$')
plt.show()
plt.plot(theta, R_para(theta), label=r'$R_\parallel$')
plt.plot(theta, T_para(theta), label=r'$T_\parallel$')
for i in theta:
  if R_para(i) < 1e-8:
    plt.plot(i,0,'ro')
    plt.text(i,0.1,r'$\theta_p$')
plt.legend()
plt.xlabel(r'$\theta$ in radians')
plt.ylabel(r'$R_\parallel and T_\parallel$')
plt.ylim(0.0,1.0)
plt.show()
```
