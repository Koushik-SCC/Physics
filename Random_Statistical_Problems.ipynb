```python
import random

N = 1000
X = [random.random() for i in range(N)]

def sigma(x,n):
    mu = sum(x)/len(x)
    A = []
    for i in x:
        A.append((abs(i - mu))**n)
    B = [j/N for j in A]
    return sum(B)/len(B)
print('sigma_1 = ', sigma(X,1))
print('sigma_2 = ', sigma(X,2))
print('sigma_3 = ', sigma(X,3))
print('sigma_4 = ', sigma(X,4))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

#(a).
R = ['r' for i in range(5)]
W = ['w' for i in range(7)]
Box = R + W
N = 5000
red = 0
for i in range(N):
    random.shuffle(Box)
    x = random.choice(Box)
    if x == 'r': red += 1
print('Probability of getting red ball: ', red/N)
```

`Theoretical Result:`

Probability of obtaining one ball as red: $ \frac{5}{5+7}=0.417 $

```python
#(b).(i)
R = ['r' for i in range(5)]
W = ['w' for i in range(7)]
Box = R + W
N = 5000
#(i).
both_red = 0
for i in range(N):
    random.shuffle(Box)
    x = random.sample(Box,2)
    if x[0] == x[1] == 'r': both_red += 1
print('Probability of getting both red ball: ', both_red/N)
```

`Theoretical Result:`

Probability of obtaining both ball as red: $ \frac{5}{12}*\frac{4}{11} = 0.152 $

```python
#(b).(ii)
R = ['r' for i in range(5)]
W = ['w' for i in range(7)]
Box = R + W
N = 5000
#(ii).
diff = 0
for i in range(N):
    random.shuffle(Box)
    x = random.sample(Box,2)
    if x[0] != x[1]: diff += 1
print('Probability of getting balls of different color: ', diff/N)
```

`Theoretical Result:`

Probability of obtaining one ball red & another as white: $ 2*\frac{5}{12}*\frac{7}{11} = 0.53 $

```python
import random
dice1 = [i for i in range(1,7)]
dice2 = [i for i in range(1,7)]
N = 5000
both3 = 0
sum7 = 0
diff = 0
for i in range(N):
    random.shuffle(dice1)
    x1 = random.choice(dice1)
    random.shuffle(dice2)
    x2 = random.choice(dice2)
    if x1 == x2 == 3: both3 += 1
    if x1 + x2 == 7 : sum7 += 1
    if x1 == 3 and x2 == 4: diff += 1
print('Probability of getting 3 on both the dices: ', both3/N)
print('Probability of getting sum of the points on the dices 7: ', sum7/N)
print('Probability of getting 3 on 1st dice  and 4 on 2nd dice: ', diff/N)
```

Two dices are completely independent

Probability of getting sum of the points 7: $ (2*\frac{3}{36}) = 0.167 $

Probability of getting 3 on 1st dice and 4 on 2nd dice: ($ \frac{1}{6}*\frac{1}{6}) = 0.0278 $

```python
import random
import matplotlib.pyplot as plt
import numpy as np

N = 10000
n = 10
while n <= 110:
    No_of_heads = np.zeros(n + 1) #0 to n + 1 (possibilities)
    for i in range(N):
        heads = 0
        for j in range(n):
            r = random.random()
            if r >= 0.5: heads += 1
        for k in range(len(No_of_heads)):
            if k == heads: No_of_heads[k] += 1
    Prob_of_no_of_heads = [P/N for P in No_of_heads]
    Heads = [i for i in range(n + 1)]

    plt.bar(Heads, Prob_of_no_of_heads, color='maroon')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability of getting Heads')
    plt.title('Number of heads ~ Probability \n Tossed '+str(n)+' times')
    plt.show()
    n += 20
```

$ \therefore $ As we increase N, the probability distribution converges to gaussian distribution

-----------------------------------------------------------------------------------------------------------------

`Comparison with theoretical result:`

$ P = \binom{N}{n}. (\frac{1}{2})^n . (\frac{1}{2})^{N-n} $

```python
import math

def P(N,n):
    Probability = (math.factorial(N)/(math.factorial(n)*math.factorial(N-n)))*(1/2)**N
    return Probability

N = 10
while N <= 110:
    Heads = np.arange(N + 1)
    Prob = []
    for n in Heads:
        Prob.append(P(N,n))
    plt.plot(Heads, Prob, color = 'darkmagenta')
    plt.title('Probability distribution of Coin Toss '+str(N)+' times')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability of getting Heads')
    plt.show()
    N += 20
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

def mean(x):
    return (sum(x)/len(x))
def variance(x):
    x_avg = sum(x)/len(x)
    x2 = [i*i for i in x]
    x2_avg = sum(x2)/len(x2)
    sigma2 = x2_avg - x_avg**2
    return sigma2
N = 1000000
X = [random.uniform(-np.pi, np.pi) for i in range(N)]

mu = mean(X)
sigma2 = variance(X)
#(a)
print('Mean: ', mu,'Variance: ', sigma2)
#(b), (c)
x_means = []
AA = []
n = 50
nn = 0
while nn < 1000:
    x_random = []
    for i in range(n):
        x_random.append(random.choice(X))
    x_mean = mean(x_random)
    x_means.append(x_mean)
    A = (x_mean - mu)/(sigma2 * n)
    AA.append(A)
    nn += 1

plt.hist(x_means, bins = 200,color='g')
plt.xlabel('Means')
plt.title('Histogram plot of means')
plt.show()
plt.hist(AA, bins = 200,color='r')
plt.title('Histogram plot of '+r'$ \frac{<X_n> - \mu}{\sigma^2 n} $')
plt.show()
```
