```python
import matplotlib.pyplot as plt
import numpy as np

A = [-1, 0, 1]
Names = ['BE', 'MB', 'FD']
def f(x,a): return 1/(np.exp(x) + a)
```

```python
x = np.arange(0.35, 3.0+0.05, 0.05)
for a in range(3):
    plt.plot(x,f(x,A[a]), label=str(Names[a]))
plt.legend()
plt.xlabel(r'$ x=\frac{\epsilon - \mu}{k_B T} $')
plt.ylabel(r'$f(\epsilon)$')
plt.title('Distribution Functions')
plt.grid(True)
plt.show()
```

```python
x1 = np.arange(-1, 3+0.05, 0.05)
x2 = np.arange(0+0.05, 3+0.05, 0.05)
for a in range(3):
    if A[a] != -1:
        plt.plot(x1,f(x1,A[a]), label=str(Names[a]))
    else:
        plt.plot(x2,f(x2,A[a]), label=str(Names[a]))
plt.legend()
plt.xlabel(r'$ x=\frac{\epsilon - \mu}{k_B T} $')
plt.ylabel(r'$f(\epsilon)$')
plt.title('Distribution Functions')
plt.grid(True)
plt.axvline(c='k',ls='--')
plt.show()
```

```python
x1 = np.arange(-3.0, 3+0.05, 0.05)
x2 = np.arange(0+0.05, 3+0.05, 0.05)
for a in range(3):
    if A[a] != -1:
        plt.plot(x1,f(x1,A[a]), label=str(Names[a]))
    else:
        plt.plot(x2,f(x2,A[a]), label=str(Names[a]))
plt.legend()
plt.xlabel(r'$ x=\frac{\epsilon - \mu}{k_B T} $')
plt.ylabel(r'$f(\epsilon)$')
plt.title('Distribution Functions')
plt.grid(True)
plt.axvline(c='k',ls='--')
plt.show()
```

let, $\frac{\epsilon}{\mu} = x$ & $ \frac{\mu}{k_BT} = y $

```python
x = np.arange(0,5.0+0.02,0.02)
def f(x,y): return 1/(np.exp((x-1)*y) + 1)
Y = [1,2,10,100]

for y in Y:
    plt.plot(x,f(x,y),label=r'$ \frac{\mu}{k_BT}= $ '+str(y))
plt.legend()
plt.grid()
plt.axhline(0.5,c='k',ls='--')
plt.title('FD Distribution')
plt.xlabel(r'$\frac{\epsilon}{\mu}$')
plt.ylabel(r'$f(\epsilon)$')
plt.show()
```
